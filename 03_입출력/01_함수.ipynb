```python
# 1. 일반적인 함수
# 입력값 O, 반환값 O
def add(a,b): #입력값 a,b
	result = a + b
	return result #반환값 return

c = add(3,5)
print(c)
```

```python
# 2. 입력값 X
def say(): # 입력값 X
	return print("hello") # 반환값 print()
```

```python
# 입력값이 몇 개가 될지 모를때는 어떻게 사용할까?

# ex) print("hi","Hello","bye")
def add_many(*args):
	result = 0
	print("args : ", args)
	print(type(args))
	for i in args:
		result = result + i
	return result
```

```python
add_many(3,4,6,7)
```
