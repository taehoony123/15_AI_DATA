```python
# 만약 3000원 이상의 돈을 가지고 있으면 택시를 타고 그렇치 않으면 걸어가라
money = 2500

if money >= 3000:
	print("택시를 탄다.")
else : 
	print("걸어간다")
```

```python
# 만약 3000원 이상의 돈을 가지고 있거나 카드가 있다면 택시를 타고 그렇치 않으면 걸어가라
money = 2000
card = True

if money >= 3000 or card:
	print("택시를 탄다.")
else : 
	print("걸어간다")
```

```python
# 만약에 주머니에 돈이 있으면 택시를 타고, 그렇지 않으면 걸어가라

pocket = ["돈", "카드"]

if "동전" in pocket :
	print("택시를 탄다.")
else :
	print("걸어간다")
```

```python
# 주머니에 돈이 있으면 택시를 타고,주머니에 돈은 없지만 카드가 있으면 택시를 타고, 그렇지 않으면 걸어가라

pocket = ["카드", "폰", "시계"]

if "돈" in pocket :
	print("택시를 탄다.")
elif "카드" in pocket :
	print("카드 택시를 탄다")
	# if "돈" in pocket에서 돈이 없으니까 elif로 내려온거기 때문에 "돈" not in pocket구절은 필요 없음
else :
	print("걸어간다")
```

```python
# match-case 문
# 학점이 A이면 탁월한 성적입니다 B이면 우수한 성적입니다 C이면 보통입니다. 그외에는 노력이 필요하다
grade = "B"

match grade:
    case "A":
        print("탁월한 성적입니다") 
    case "B":
        print("우수한 성적입니다") 
    case "C":
        print("보통입니다") 
    case _:
        print("노력이 필요합니다")
```

```python
grade = "B"

if grade == "A" :
	print("탁월한 성적입니다")
elif grade == "B" :
	print("우수한 성적입니다")
elif grade == "C" :
	print("보통입니다")
else :
	print("노력이 필요합니다")
```

```python
# 사용자로 부터 숫자를 입력 받는다
# 0이상 100 이하의 숫자만 받는다 
# 19 이상이면 성인입니다
# 그렇지 않으면 미성년자 입니다.

age = int(input("나이를 입력해주세요"))

if 0<= age <= 100 :
	if age >= 19 :
		print("성인입니다.")
	else :
		print("미성년자 입니다.")
else : 
	print("0에서 100 사이의 숫자를 입력해주세요")


```
