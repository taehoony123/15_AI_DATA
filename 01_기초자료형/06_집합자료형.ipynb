```python
# 집합자료형은 어떻게 만들까?
# 리스트 [] 튜플 () 딕셔너리 { : } 집합 {}
s1 = set([1,2,3])
s2 = {1,2,3}

print(s1)
print(type(s1))

print(s2)
print(type(s2))
```

```python
# 집합자료형의 특징
# 1. 중복허용 안된다.
# 2. 순서가 없다.
```

```python
#리스트 lst를 set함수에 넣으면 중복이 제거된다.
lst = [1,2,3,1,2,3,1,2]
s1 = set(lst)

print(s1)
# s1을 다시 리스트로 변환하기
s2 = list(s1)
print(s1)
```

```python
# 교집합 구하기
s1 = {1,2,3,4,5,6}
s2 = {4,5,6,7,8,9}

print(s1.intersection(s2))
print(s1 & s2) #intersection

# 합집합
print(s1.union(s2))
print(s1 | s2)

# 차집합
print(s1.difference(s2))
print(s1 - s2)
print(s2 - s1)

```
