```python
# 딕셔너리는 어떻게 만들까
# {key1: value1, key2 : value2}

dic = {
	"name" : "ken",
	"phone" : "010-****-****",
	"birth" : "01/01",
}

print(dic)
print(type(dic))
```

```python
# 딕셔너리 쌍 추가하기
a = {1: "a"}
a[2] = "b"
print(a)
a["hobby"] = ["running", "football"]
print(a)
```

```python
# 딕셔너리 삭제
del a[1]
print(a)
del a["hobby"]
print(a)
```

```python
# 딕셔너리에서 Key를 사용해 Value값 얻기

data = {"유재석" : "개그맨",
		"손흥민" : "축구선수",
		"귀도" : "개발자"}

print(type(data))

print(data["유재석"])

data["귀도"] = "코딩"
print(data)

data["이재용"] = "CEO"
print(data)
```

```python
# 딕셔너리를 만들때 주의사항
# 1.딕셔너리에서 Key는 고유한 값이므로 중복이 안된다.
# 2.수정이 되는 데이터 타입을 key로 올 수 없다. -> List
a = {
	[1,2] : "hi"
}
```

```python
a = {
	1: "a",
	"a" : 10
}
print(a)

# 1 <-- "a" --> 10
# a가 변경 된게 아니라 1을 가리키고 있다가 10을 가리키는 걸로 바뀌는 의미이다. 변경의 의미는 아님!
# 리스트는 [1,2,3,4] -> [1,2,3,5] 이건 변경의 의미! 그래서 딕셔너리에서 list 불가
```

```python
# 딕셔너리 관련 함수
# Key 리스트 만들기 - keys

data = {"유재석" : "개그맨",
		"손흥민" : "축구선수",
		"귀도" : "개발자"}

# key와 value 값 가지고 오기 (함수이기 때문에 ()를 keys() 이렇게 넣어줘야함)
print(data.keys())
print(data.values())

# list로 key와 value 가지고오기
print(list(data.keys())) 
print(list(data.values()))

# key와 value 쌍 얻기
print(data.items())
print(list(data.items())) 

# key와 value 쌍 모두 지우기
print(data.clear())
print(data)

```

```python
# key로 value 값 얻기
data = {"유재석" : "개그맨",
		"손흥민" : "축구선수",
		"귀도" : "개발자"}

print(data.get("유재석"))
print(data.get("김연아"))
print(data["손흥민"])
print(data["김연아"])
```

```python
# 해당 key가 딕셔너리 안에 있는지 조사하기 - in
print("유재석" in data)
print("김연아" in data)
```

```python
# key로 value 꺼내기 - pop
data = {"유재석" : "개그맨",
		"손흥민" : "축구선수",
		"귀도" : "개발자"}

poped = data.pop("유재석")
print(poped)
print(data)
```
