```python
# 숫자형은 어떻게 사용할까? 
## 정수형 (-1, 0 ,1 (int))

a = 123
b = -11
c = 0

# shift + alt 문장복사 
print(a)
print(b)
print(c)

print(type(a))
print(type(b))
print(type(c))
```

```python
## 실수형
# E5 = 10의 5제곱
a = 1.2
b = -3.14
c = 4.2E5
d = 4.2E-5

print(a)
print(b)
print(c)
print(d)

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

```python
## 8진수 16진수
# 우리가 쓰는건 10진수
# 예를 들어 314는 3*(10**2) + 1*(10**1) + 4
#  314는 3*(8**2) + 1*(8**1) + 4

exam = 3*(8**2) + 1*(8**1) + 4
# 8진수
a = 0o314 
b = 0o10
c = 0o7

print("exam = ",exam)
print("a = ", a)
print("b = ", b)
print("c = ", c)

print("===============================")
# 16진수
d = 0x8ff
e = 0xabc
f = 0x17

print("d = ", d)
print("e = ", e)
print("f = ", f)
```

```python
# 사칙연산 (+, - , * , /)
a = 3
b = 4

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

```python
# x의 y제곱을 나타내는 **연산자
a = 3
b = 4

print(a ** b)
```

```python
# 나눗셈 후 나머지 % 연산자 
a = 7
b = 3

print(a % b)
```

```python
# 나눗셈 후 몫을 반환하는 // 연산자
a = 7
b = 3

print(a // b)
```

```python
# 복합 연산자 
a = 1
a += 1  # a = a + 1 

print(a)

# 곱하기 제곱 나누기 나머지 등등 다 가능함
```
