```python
# 문자열은 어떻게 만들고 사용할까

# 1. 큰따옴표로 사용하기
print("Hello world")

# 2. 작은따옴표로 사용하기
print('Python is fun')

# 3. 큰따옴표 3개를 사용 
print("""Life is too short, 
	  you need python """)

# 4. 작은따옴표 3개 사용
print('''Life is too short, 
	  you need python''')
```

```python
print("Python's favorate is apple")
# 여기서 작은따옴표는 문자열로 해당하고 큰 따옴표는 명령하는 의미이다.

print('Python is very easy, "he says"')
```

```python
# 역슬래시 \ 를 이용하여 작은따옴표와 큰따옴표를 문자열에 포함
print('Python\'s favorate is apple')
print("\"Python is very easy.\" he says")
```

```python
# 역슬래시 \n 는 줄 바꿈
print("Life is too short,\nyou need python \n")

# 역슬래시 \t 는 탭이다
print("Life is too short,\tyou need python \n")

# 역슬래시 \\ 는 \
print("Life is too short,\\ you need python ")
```

```python
# 문자열 연산
## 문자열 더해서 연결하기
a = "python "
b = "is fun"

print(a + b)

## 문자열 곱하기 
a = "python "
print(a * 3)
```

```python
# 문자열 길이 구하기
a = "Life is too short, you need python"
len(a)
```

```python
# 문자열 인덱싱
a = "Life is too short, you need python"
print(a[0])
print(a[5])
print(a[8])
print(a[-1])
print()
# 문자열 슬라이싱
print(a[:4])
print(a[-6:])
print(a[8:11])
```

```python
# 문자열 포맷팅 따라하기
# 1. 숫자 바로 대입
print("I eat %d apples" % 3)

# 2. 문자열 바로 대입
print("i eat %s apples" %"five")

# 3. 숫자값을 나타내는 변수로 대입
number = 3
print("i eat %d apples" % number)

# 4. 두개 이상의 값을 대입
number = 10
day = "three"
print("i ate %d apples, so i was sick for %s days" % (number,day))
```

```python
# 문자열 포맷 코드
# %s = 문자열 
# %d = 정수
# %f = 실수
# %o = 8진수
# %x = 16진수
# %c = 문자형
```

```python
# 정렬과 공백
print("%10s" % "hi")
```

```python
# 소수점 표현하기
print("%10.4f" % 3.4213234)
```

```python
# 숫자 바로 대입하기
print("I eat {0} apples".format(3))

# 문자열 바로 대입하기
print("I eat {0} apples".format("five"))

# 2개 이상의 값 넣기
number = 10
day = "three"
print("I ate {0} apples. so I was sick for {1} days.".format(number,day))
```

```python
# f문자열 포맷팅
name = "홍길동"
age = "30"
print(f"나의 이름은 {name}이고 나이는 {age}세 입니다.")
```

```python
# 문자열 관련 함수들
# 1. 문자 개수 세기 = count

a = "hobby"
a.count("b")
```

```python
# 2. 위치 알려주기 함수 = find
a = "python is the best choice"
a.find("b")
```

```python
# 3. 위치 알려주기 함수2 = index
a = "life is too short"
a.index("t")
```

```python
# 5. 소문자를 대문자로 바꾸는 함수 - upper
a = "hi"
a.upper()
```

```python
# 6. 대문자를 소문자로 - lower
a = "HI"
a.lower()
```

```python
# 7. 왼쪽 공백 지우기 - lstrip
a = "   hi   "
a.lstrip()
```

```python
# 8. 오른쪽 공백 지우기 - rstrip
a = "   hi   "
a.rstrip()
```

```python
# 9. 양쪽 공백 지우기 - strip
a= "   hi   "
a.strip()
```

```python
# 10. 문자열 바꾸기 - replace
a = "life is too short"
a.replace("life","leg")
```
