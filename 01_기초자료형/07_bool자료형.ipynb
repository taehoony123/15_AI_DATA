```python
# 불 자료형은 어떻게 사용할까?
a = True
b = False

print(a,b)
print(type(a))
print(type(b))
```

```python
# 같다
print(1 == 1)
print(1 == 2)
# 부등호
print(2 > 1)
print(1 > 2)
print(2 >= 1)
print(2 <= 1)
# 같지 않다
print(1 != 1)
```

```python
# 자료형의 참과 거짓
print(bool("python" ))
print(bool(2))
print("=" * 10)

# 공백은 True
print(bool(" "))
print(bool(""))
print("=" * 10)

# List
print(bool([]))
print(bool(list[1,2]))
print("=" * 10)

# Tuple
print(bool(()))
print(bool((1,)))
print("=" * 10)
# set
print(bool(set([])))
print(bool(set([1,2])))
```

```python
# 논리 연산자

# and 연산자
print(True and True)
print(True and False)
print("=" * 10)
# or 연산자
print(True or True)
print(True or False)
print("=" * 10)
# not 연산자
print(not True)
print(not 1)
```

```python
# ex)
x = 5
y = 10
print(x > 0 and y > 0)
print(x != y)
```
