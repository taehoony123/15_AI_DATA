### 문제 7: 소수 판별기 
사용자가 입력한 숫자가 소수인지 판별하고, 1부터 해당 숫자까지의 모든 소수를 출력하는 프로그램을 작성하세요.

**요구사항**:
- for 문과 if 문 사용
- 2 이상의 숫자만 입력 가능
- 소수 판별 로직 구현

**실행 예시**:
```
숫자를 입력하세요: 20
20은(는) 소수가 아닙니다.

1부터 20까지의 소수:
2, 3, 5, 7, 11, 13, 17, 19
총 8개의 소수가 있습니다.
```

````python
num = int(input("숫자를 입력하세요: "))
primes = []

is_prime = True # 이건 소수다!
if num < 2: is_prime = False # 너무 적으면 소수가 아니다. 0,1

for i in range(2, int(num**0.5) + 1): #2,3 
    if num % i == 0: #나누어 떨어지면 탈락 (소수 아님)
        is_prime = False
        break



print(f"{num}은(는) {'소수입니다.' if is_prime else '소수가 아닙니다.'}")

for i in range(2, num + 1):
    check = True
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            check = False
            break
    if check: primes.append(i)

print(f"\n1부터 {num}까지의 소수:\n{', '.join(map(str, primes))}")
print(f"총 {len(primes)}개의 소수가 있습니다.")
````

````python
import random

target = random.randint(1, 100)
count = 0
print("1부터 100사이의 숫자를 맞춰보세요!")

while True:
    count += 1
    guess = int(input(f"시도 {count}: "))
    
    if guess < target:
        print("Up!")
    elif guess > target:
        print("Down!")
    else:
        print(f"정답입니다! {count}번 만에 맞추셨습니다!")
````

````python
products = {"1.사과": 1000, "2.우유": 2500, "3.고기": 15000, "4.빵": 2000, "5.커피": 4000}
cart = []

while True:
    menu = input("\n1.목록 2.담기 3.보기 4.결제 5.종료: ")
    if menu == '1':
        for p, price in products.items(): print(f"{p}: {price}원")
    elif menu == '2':
        item = input("상품명 입력: ")
        for p in products:
            if item in p: 
                cart.append((p, products[p]))
                print("담기 완료!")
    elif menu == '3':
        for i in cart: print(f"- {i[0]}: {i[1]}원")
    elif menu == '4':
        total = sum(item[1] for item in cart)
        coupon = input("할인쿠폰이 있나요? (y/n): ")
        if coupon == 'y': total *= 0.9
        ship = 0 if total >= 30000 else 3000
        print(f"총 결제 금액: {total + ship:,.0f}원 (배송비: {ship}원)")
        break
    elif menu == '5': break
````
